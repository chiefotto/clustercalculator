```python
cleaned_schedule = full_schedule[['leagueId', 'seasonYear', 'gameId', 'gameDateEst', 'gameTimeEst', 'weekNumber', 'monthNum', 'homeTeam_teamId', 'awayTeam_teamId']]
pd.DataFrame(cleaned_schedule)
cleaned_schedule['gameDateEst'] = pd.to_datetime(cleaned_schedule['gameDateEst']).dt.date
cleaned_schedule
```

```python
# Convert column to date objects
cleaned_schedule['gameDateEst'] = pd.to_datetime(cleaned_schedule['gameDateEst']).dt.date

# Convert today to a date object
today_date = pd.Timestamp.today().date()

# Now the filter will work
# todays_schedule = cleaned_schedule[cleaned_schedule['gameDateEst'] == today_date]
todays_schedule = full_schedule.loc[full_schedule["gameDateEst"] == today_date].copy(deep=True)

todays_schedule
```

```python
clusters = pd.read_parquet('../data/jan21clusters.parquet')

clusters = clusters[["TEAM_ID", "cluster"]]
clusters
```

```python
from nba_api.stats.static import teams as nba_teams
teams = nba_teams.get_teams()
teams_df = pd.DataFrame(teams)
```

```python
teams_revised = teams_df[['full_name', 'id', 'abbreviation']]
teams_revised
```

```python
merged_df = pd.merge(clusters, teams_revised, left_on='TEAM_ID', right_on='id').drop(columns=['id'])
```

```python
merged_df = merged_df.copy(deep=True)
merged_df
```

```python
# Create two mapping dictionaries
cluster_map = dict(zip(merged_df['TEAM_ID'], merged_df['cluster']))
name_map = dict(zip(merged_df['TEAM_ID'], merged_df['full_name']))

# Map them to your schedule
todays_schedule['home_team_cluster'] = todays_schedule['homeTeam_teamId'].map(cluster_map)
todays_schedule['away_team_cluster'] = todays_schedule['awayTeam_teamId'].map(cluster_map)

todays_schedule['home_team_name'] = todays_schedule['homeTeam_teamId'].map(name_map)
todays_schedule['away_team_name'] = todays_schedule['awayTeam_teamId'].map(name_map)

```
